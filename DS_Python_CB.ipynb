```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Graficando una función lineal
def f(m, x, b):
    return m*x + b

res = 100 #Cantidad de puntos a gráficar
b = 5
m = 10
x = np.linspace(-10, 10, num = res)
y = f(m, x, b)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```python
#Graficando una parábola

def f(x):
    return 12*x*x + 3*x + 121

res = 100 #Cantidad de puntos a gráficar
x = np.linspace(-10, 10, num = res)
y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```python
#Graficando varias funciones trascendales

def f(x):
    return np.cos(x)

def g(x):
    return np.sin(x)



res = 100 #Cantidad de puntos a gráficar
x = np.linspace(-10, 10, num = res)
y = f(x)
z = g(x)


fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].set_title('Coseno')
ax[1].set_title('Seno')
ax[0].plot(x, y)
ax[0].grid()
ax[1].plot(x, z)
ax[1].grid()
ax[0].axhline(y = 0, color = 'r')
ax[1].axhline(y = 0, color = 'r')
ax[0].axvline(x = 0, color = 'r')
ax[1].axvline(x = 0, color = 'r')

```

```python
#Descenso del Gradiente
from matplotlib import cm
import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})

def f(x,y):
    return x**2 + y**2

x = np.linspace(-4, 4, num = 100)
y = np.linspace(-4, 4, num = 100)

X,Y = np.meshgrid(x,y)

Z = f(X,Y)
surf = ax.plot_surface(X, Y, Z, cmap = cm.cool)
fig.colorbar(surf)
```

```python
#Curvas de nivel
level_map = np.linspace(np.min(Z), np.max(Z), num = 100)
plt.contourf(X, Y, Z, levels= level_map, cmap = cm.cool)
plt.colorbar()
plt.title('Descenso del gradiente')
#Aplicando el algoritmo del descenso del gradiente
p = np.random.rand(2) * 8 - 4
plt.plot(p[0], p[1], 'o', c ='k')
h = 0.001
lr = 0.01

def derivate(cp, p):
    return (f(cp[0], cp[1]) - f(p[0], p[1]))/h


def gradient(p):
    grad = np.zeros(2)
    for idx, val in enumerate(p):
        cp = np.copy(p)
        cp[idx] = cp[idx] + h
        dp = derivate(cp, p)
        grad[idx] = dp
    
    return grad

for i in range(100):
    p = p - lr*gradient(p)
    if i%10 == 0:
        plt.plot(p[0], p[1], 'o', c = 'r')

plt.plot(p[0], p[1], 'o', c = 'w')
```
